```python
def nextstep():
    a = random.randrange(0,100)
    if a >= 50:
        return 1
    else:
        return -1
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

x = [0]
y = [0]
a = [0]
b = [0]


fig, ax = plt.subplots()
line1, = ax.plot(x, y, color = "r")
line2, = ax.plot(a, b, color = "g")

def update(num, x, y, z, line1, line2):
    x.append(x[-1]+nextstep())
    y.append(y[-1]+nextstep())
    a.append(a[-1]+nextstep())
    b.append(b[-1]+nextstep())
    line1.set_data(x, y)
    line2.set_data(a, b)
    return [line1,line2]

ani = animation.FuncAnimation(fig, update, len(x), fargs=[x, y, z, line1, line2],
                  interval=200, blit=False)

ax.set_xlabel('Age (day)')
ax.set_ylabel('EO (%)')

plt.show()
```
