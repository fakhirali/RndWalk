```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
import random


def nextstep():
    rnd = random.randrange(0,100)
    if rnd >= 50:
        return 1
    else:
        return -1

def init():
    sz = 50
    ax.set_ylim(-sz, sz)
    ax.set_xlim(-sz, sz)
    line.set_data(xdata, ydata)
    return line,

fig, ax = plt.subplots()
line, = ax.plot([], [], lw=2)
ax.grid()
xdata, ydata = [0], [0]


def run(frame):
    xdata.append(xdata[-1]+nextstep())
    ydata.append(ydata[-1]+nextstep())
    line.set_data(xdata, ydata)
    return line,

ani = animation.FuncAnimation(fig, run, frames=None, blit=True, interval=20,
                              repeat=False, init_func=init)
plt.show()
```
